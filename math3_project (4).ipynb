# Euler

```python
# y = 4y + 4x + 1
# x0 = 0
# y0 = 0
# h = 0.1
```

```python
#------------------Input------------------ 
x0 = 0       #input
y0 = 0       #input
h = 0.1      #input
max = 20      #input   
#------------------Solution------------------ 
y_list = []
x_list = []
y_list.append(y0)
x_list.append(x0)
for i in range(0,max):
  xx = x_list[-1]
  yy = y_list[-1]
  y = yy +( h * ( (4 * yy) + (4 * xx) + 1) )
  x = xx + h
  y = round(y , 3)
  x = round(x , 3)
  y_list.append(y)
  x_list.append(x)
#------------------answer visualization------------------ 
data = {"x" : x_list , "y" : y_list}
df_euler = pd.DataFrame(data)
df_euler
```

# improved euler

```python
# y = 4y + 4x + 1
# x0 = 0
# y0 = 0
# h = 0.1
#------------------Input------------------ 
x0_ie = 0       #input
y0_ie = 0       #input
h_ie = 0.1      #input
max_ie = 20      #input
#------------------Solution------------------ 
y_list_ie = [y0_ie]
x_list_ie = [x0_ie]
for i in range(0,max_ie):
  yie = y_list_ie[-1]
  xie = x_list_ie[-1]
  x = xie + h_ie  
  k1 = h_ie * ((4*yie) + (4*xie) + 1)
  k2 = h_ie * ((4*(yie + k1)) + (4*(x)) + 1)
  y = yie + (0.5*(k1+k2))
  y_list_ie.append(y)
  x_list_ie.append(x)
#------------------answer visualization------------------ 
dataie = {"x" : x_list_ie , "y" : y_list_ie}
df_ieuler = pd.DataFrame(dataie)
df_ieuler
```

# runge kutta


```python
# y = 4y + 4x + 1
# x0 = 0
# y0 = 0
# h = 0.1
#------------------Input------------------ 
x0_rg = 0       #input
y0_rg = 0       #input
h_rg = 0.1      #input
max_rg = 20      #input
#------------------Solution------------------ 
y_list_rg = [y0_rg]
x_list_rg = [x0_rg]
for i in range(0,max_rg):
  yrg = y_list_rg[-1]
  xrg = x_list_rg[-1]
  k1 =  h_rg * ((4 * yrg) + (4 * xrg) + 1)
  k2 =  h_rg * ((4 * (yrg + (0.5 * k1))) + (4 * (xrg + (0.5 * h_rg))) + 1)
  k3 =  h_rg * ((4 * (yrg + (0.5 * k2)) ) + (4 * (xrg + (0.5 * h_rg )) + 1))
  k4 =  h_rg * ((4 * (yrg + k3)) + (4 * (xrg + h_rg)) + 1)
  x = xrg + h_rg
  y = yrg + ((1/6) * (k1 + (2*k2) + (2*k3) + k4) )
  y = round(y , 6)
  x = round(x , 6)
  y_list_rg.append(y)
  x_list_rg.append(x)
#------------------answer visualization------------------ 
datarg = {"x" : x_list_rg , "y" : y_list_rg}
df_runge = pd.DataFrame(datarg)
df_runge
```
